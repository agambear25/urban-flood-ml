```python
import ee
ee.Initialize()

W, S, E, N = 76.8, 28.3, 77.6, 28.9
AOI = ee.Geometry.Rectangle([W, S, E, N])
FOLDER = "maven_rasters"

def export(image, name, scale):
    task = ee.batch.Export.image.toDrive(
        image=image.clip(AOI).toFloat(),
        description=name, folder=FOLDER,
        fileNamePrefix=name, region=AOI,
        scale=scale, crs="EPSG:4326", maxPixels=int(1e10),
    )
    task.start()
    print(f"started: {name} → Drive/{FOLDER}/{name}.tif")

dw = ee.ImageCollection("GOOGLE/DYNAMICWORLD/V1").filterBounds(AOI).filterDate("2025-12-01","2026-06-01")
export(dw.select("built").median(), "built", 10)

s2 = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED").filterBounds(AOI).filterDate("2025-10-01","2026-03-01").filter(ee.Filter.lt("CLOUDY_PIXEL_PERCENTAGE", 35))
export(s2.median().normalizedDifference(["B8","B4"]).rename("ndvi"), "green", 10)

ntl = ee.ImageCollection("NOAA/VIIRS/DNB/MONTHLY_V1/VCMSLCFG").filterBounds(AOI).filterDate("2025-06-01","2026-06-01").select("avg_rad")
export(ntl.median(), "nightlights", 500)

pop = ee.ImageCollection("JRC/GHSL/P2023A/GHS_POP").filterDate("2025-01-01","2026-01-01").first().select("population_count")
export(pop, "population", 100)
```
